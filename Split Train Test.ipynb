## Libraries

```python
import pandas as pd
```

## Load the dataframes

```python
df_base = pd.read_pickle('df_base.pkl')
df_post_covid = pd.read_pickle('df_post_covid.pkl')
```

## Split Train/Test 80/20
We do not do a random split since we want the model to predict future prices based on past data. Hence, train data will be earlier than test data.

```python
# 1. Sort by time column
df_base_sorted = df_base.sort_values(by='months_since_base')
df_post_sorted = df_post_covid.sort_values(by='months_since_base')

# 2. Compute split indices
split_idx_base = int(len(df_base_sorted) * 0.8)
split_idx_post = int(len(df_post_sorted) * 0.8)

# 3. Split df_base into train/test
df_base_train = df_base_sorted.iloc[:split_idx_base]
df_base_test = df_base_sorted.iloc[split_idx_base:]

X_base_train = df_base_train.drop(columns=['log_resale_price'])
y_base_train = df_base_train['log_resale_price']
X_base_test = df_base_test.drop(columns=['log_resale_price'])
y_base_test = df_base_test['log_resale_price']

# 4. Split df_post_covid into train/test
df_post_train = df_post_sorted.iloc[:split_idx_post]
df_post_test = df_post_sorted.iloc[split_idx_post:]

X_post_train = df_post_train.drop(columns=['log_resale_price'])
y_post_train = df_post_train['log_resale_price']
X_post_test = df_post_test.drop(columns=['log_resale_price'])
y_post_test = df_post_test['log_resale_price']
```

## Save files

```python
# 5. Save all 8 files
X_base_train.to_pickle("X_base_train.pkl")
X_base_test.to_pickle("X_base_test.pkl")
y_base_train.to_pickle("y_base_train.pkl")
y_base_test.to_pickle("y_base_test.pkl")

X_post_train.to_pickle("X_post_train.pkl")
X_post_test.to_pickle("X_post_test.pkl")
y_post_train.to_pickle("y_post_train.pkl")
y_post_test.to_pickle("y_post_test.pkl")
```
